```python
import re
import matplotlib.pyplot as plt

#   
log_file_path = 'train_005930_a3c_lstm.log'

#    
epochs = []
pvs = []

#   
with open(log_file_path, 'r') as file:
    for line in file:
        #      
        match = re.search(r'\[Epoch (\d+)/\d+\] .* PV:([\d,]+)', line)
        if match:
            epoch = int(match.group(1))
            pv = int(match.group(2).replace(',', ''))  #     
            
            epochs.append(epoch)
            pvs.append(pv)

#  
plt.figure(figsize=(10, 5))
plt.plot(epochs, pvs, label='PV', color='green')
plt.title('PV over Epochs')
plt.xlabel('Epochs')
plt.ylabel('PV')
plt.grid()
plt.legend()

#  
plt.tight_layout()
plt.show()
```

```python

import re
import matplotlib.pyplot as plt

#   
log_file_path = 'train_train010_ddpg_lstm.log'

#    
epochs = []
loss = []

#   
with open(log_file_path, 'r') as file:
    for line in file:
        #      
        match = re.search(r'\[Epoch (\d+)/\d+\] .* Loss:([\d,]+)', line)
        if match:
            epoch = int(match.group(1))
            l = float(match.group(2).replace(',', ''))  #     
            
            epochs.append(epoch)
            loss.append(l)

#  
plt.figure(figsize=(10, 5))
plt.plot(epochs, loss, label='Loss', color='green')
plt.title('Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid()
plt.legend()

#  
plt.tight_layout()
plt.savefig('train_train010_ddpg_lstm_loss.png')
plt.show()
```
